5. (9б). Створіть клас 3D-фігура без атрибутів, з методом get_volume для отримання об'єму фігури, що повертає 0 та __add __, який приймає self та other в якості аргументів, а повертає суму об'ємів 3d-фігур self та other.

6. (11б). Створіть 2 дочірніх класи від 3d-фігура: Циліндр та Куб, які мають атрибути, необхідні для розрахунку об'єму. Визначте метод get_volume в кожному з них так, щоби вони розраховували об'єм у залежності від формули для кожного типу фігури. Створіть обʼєкт класу Циліндр та обʼєкт класу Куб. Виконайте операцію суми за допомогою оператора + між ними, аби знайти сумарний об'єм.

```python
import math

class Shape3D:
    def get_volume(self):
        return 0
    def __add__(self, other):
        return self.get_volume() + other.get_volume()

class Cylinder(Shape3D):
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height
    def get_volume(self):
        return math.pi * self.radius**2 * self.height

class Cube(Shape3D):
    def __init__(self, side):
        self.side = side
    def get_volume(self):
        return self.side**3

cyl = Cylinder(2, 5)
cube = Cube(3)
print("Об'єм циліндра:", cyl.get_volume())
print("Об'єм куба:", cube.get_volume())
print("Сумарний об'єм:", cyl + cube)
```

8. (13б). Створіть клас EWallet з приватними атрибутами balance та id. Реалізуйте методи поповнення та зняття коштів, забезпечивши належну інкапсуляцію. Підказка: використовуйте декоратори property та setter.

```python
class EWallet:
    def __init__(self, wallet_id, balance=0):
        self.__id = wallet_id
        self.__balance = balance

    @property
    def balance(self):
        return self.__balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Поповнено {amount}. Новий баланс: {self.__balance}")
        else:
            print("Сума поповнення має бути додатньою")

    def withdraw(self, amount):
        if amount > 0 and amount <= self.__balance:
            self.__balance -= amount
            print(f"Знято {amount}. Новий баланс: {self.__balance}")
        else:
            print("Недостатньо коштів або неправильна сума")

    @property
    def id(self):
        return self.__id

wallet = EWallet("23765", 100)
print("Баланс:", wallet.balance)
wallet.deposit(70)
wallet.withdraw(12)
print("ID гаманця:", wallet.id)
```

9. (11б). Створіть клас ShoppingCart, який містить список об'єктів типу Item. Реалізуйте методи для додавання айтемів, їх видалення та відображення списку айтемів.

```python
class Item:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def __str__(self):
        return f"{self.name} - ${self.price}"

class ShoppingCart:
    def __init__(self):
        self.items = []

    def add_item(self, item):
        self.items.append(item)
        print(f"{item.name} додано до кошика")

    def remove_item(self, item_name):
        for item in self.items:
            if item.name == item_name:
                self.items.remove(item)
                print(f"{item.name} видалено з кошика")
                return
        print(f"{item_name} не знайдено в кошику")

    def show_items(self):
        if not self.items:
            print("Кошик порожній")
        else:
            print("Ваш кошик:")
            for item in self.items:
                print(item)

cart = ShoppingCart()
item1 = Item("Яблуко", 25)
item2 = Item("Банан", 15)

cart.add_item(item1)
cart.add_item(item2)
cart.show_items()

cart.remove_item("Яблуко")
cart.show_items()
```

Створіть клас Car, який успадковує обидва класи Vehicle та Engine. Продемонструйте використання множинної спадковості, створивши обʼєкт класу Car та отримавши mro для цього класу.

```python
class Vehicle:
    def __init__(self, brand, production_age):
        self.brand = brand
        self.production_age = production_age

class Engine:
    def __init__(self, engine_type, size):
        self.engine_type = engine_type
        self.size = size

class Car(Vehicle, Engine):
    def __init__(self, brand, production_age, engine_type, size):
        Vehicle.__init__(self, brand, production_age)
        Engine.__init__(self, engine_type, size)

my_car = Car("Toyota", 5, "Petrol", 2.0)
print("Бренд:", my_car.brand)
print("Вік виробництва:", my_car.production_age)
print("Тип двигуна:", my_car.engine_type)
print("Об'єм двигуна:", my_car.size, "л")
print("MRO класу Car:", Car.mro())
```

Ітерабельний об’єкт — це той, по якому можна пройтися (рядок або список), а ітератор — це спеціальний об’єкт, який реально перебирає елементи по одному за допомогою next().

13. (11б). Створіть клас, що має визначений Iterator Protocol та при кожному виклику next повертає назву дня тижня, поки вони не вичерпаються. Зверніть увагу, що порядок має значення.

```python
class WeekDays:
    def __init__(self):
        self.days = [
            "Monday", "Tuesday", "Wednesday",
            "Thursday", "Friday", "Saturday", "Sunday"
        ]
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index < len(self.days):
            day = self.days[self.index]
            self.index += 1
            return day
        else:
            raise StopIteration

week = WeekDays()
for day in week:
    print(day)
```
